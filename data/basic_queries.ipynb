```python
import duckdb
```

```python
con = duckdb.connect("chocolate_sales.db")
```

```python
df_calendar = con.execute("SELECT * FROM chocolate_sales.calendar").df()
print(df_calendar)
```

```python
df_customers = con.execute("SELECT * FROM chocolate_sales.customers").df()
print(df_customers)
```

```python
df_products = con.execute("SELECT * FROM chocolate_sales.products").df()
print(df_products)
```

```python
df_sales = con.execute("SELECT * FROM chocolate_sales.sales").df()
print(df_sales)
```

```python
df_stores = con.execute("SELECT * FROM chocolate_sales.stores").df()
print(df_stores)
```
